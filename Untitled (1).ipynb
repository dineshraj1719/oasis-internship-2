```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("car data.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()

df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(df['Fuel_Type'])

plt.title("Fuel Type Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(df['Transmission'])

plt.title("Transmission Type")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(df['Selling_type'])

plt.title("Selling Type")

plt.show()
```

```python
df['Current_Year'] = 2024

df['No_Years'] = df['Current_Year'] - df['Year']
```

```python
df.drop(['Car_Name','Year','Current_Year'],
        axis=1,
        inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(
    df,
    drop_first=True
)
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```

```python
X = df.drop('Selling_Price',
            axis=1)

y = df['Selling_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
lr = LinearRegression()

lr.fit(X_train,
       y_train)

pred_lr = lr.predict(X_test)

print("R2 Score:",
      r2_score(y_test,pred_lr))
```

```python
dt = DecisionTreeRegressor(
    random_state=42
)

dt.fit(X_train,
       y_train)

pred_dt = dt.predict(X_test)

print("R2 Score:",
      r2_score(y_test,pred_dt))
```

```python
rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train,
       y_train)

pred_rf = rf.predict(X_test)

print("R2 Score:",
      r2_score(y_test,pred_rf))
```

```python
results = pd.DataFrame({
    'Model':
    ['Linear Regression',
     'Decision Tree',
     'Random Forest'],

    'R2 Score':
    [
     r2_score(y_test,pred_lr),
     r2_score(y_test,pred_dt),
     r2_score(y_test,pred_rf)
    ]
})

results
```

```python
sns.barplot(
    x='R2 Score',
    y='Model',
    data=results
)

plt.title("Model Comparison")

plt.show()
```

```python
print("MAE:",
      mean_absolute_error(y_test,pred_rf))

print("MSE:",
      mean_squared_error(y_test,pred_rf))

print("RMSE:",
      np.sqrt(
          mean_squared_error(
              y_test,
              pred_rf
          )
      ))
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    y_test,
    pred_rf
)

plt.xlabel("Actual Price")

plt.ylabel("Predicted Price")

plt.title("Actual vs Predicted")

plt.show()
```

```python
params = {
    'n_estimators':[100,200,300],
    'max_depth':[5,10,15],
    'min_samples_split':[2,5,10]
}

grid = GridSearchCV(
    RandomForestRegressor(
        random_state=42
    ),
    params,
    cv=5,
    scoring='r2'
)

grid.fit(X_train,
         y_train)

print(grid.best_params_)
```

```python
best_rf = grid.best_estimator_

pred_best = best_rf.predict(X_test)

print("Optimized R2 Score:",
      r2_score(y_test,pred_best))
```

```python
importance = pd.Series(
    best_rf.feature_importances_,
    index=X.columns
)

importance.sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.title(
    "Feature Importance"
)

plt.show()
```

```python
scores = cross_val_score(
    best_rf,
    X,
    y,
    cv=5,
    scoring='r2'
)

print(scores)

print("Average Score:",
      scores.mean())
```

```python
import pickle

file = open(
    "car_price_model.pkl",
    "wb"
)

pickle.dump(
    best_rf,
    file
)

file.close()
```
